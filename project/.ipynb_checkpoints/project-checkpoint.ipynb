```python
import cv2
import numpy as np

def resize(img):
    IMAGE_H = 680
    IMAGE_W = 640
    dim = (IMAGE_W, IMAGE_H)
    resized = cv2.resize(img, dim)
    return resized

def resize1(img):
    IMAGE_H = 640
    IMAGE_W = 680
    dim = (IMAGE_W, IMAGE_H)
    resized = cv2.resize(img, dim)
    return resized

def rotate180(img):
    img_rotate_180 = cv2.rotate(img, cv2.ROTATE_180)
    return img_rotate_180

def rotate90counterclockwise(img):
    img_rotate_90_counterclockwise = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
    return img_rotate_90_counterclockwise

def rotate90clockwise(img):
    img_rotate_90_clockwise = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
    return img_rotate_90_clockwise
    

def sketch(img):
    IMAGE_H = 680
    IMAGE_W = 640

    src = np.float32([[0, 0], [0, IMAGE_H], [IMAGE_W, IMAGE_H], [IMAGE_W, 0]])
    dst = np.float32([[int(0.296875*IMAGE_W),197], [int(0.296875*IMAGE_W), 482], 
                      [int(0.703125*IMAGE_W), 482], [int(0.703125*IMAGE_W), 197]])
    M = cv2.getPerspectiveTransform(src, dst) # The transformation matrix
    
    #img = img[0:IMAGE_H, 0:IMAGE_W] # Apply np slicing for ROI crop
    warped_img = cv2.warpPerspective(img, M, (IMAGE_W, IMAGE_H)) # Image warping
    
    return warped_img

def sketch1(img):
    IMAGE_H = 680
    IMAGE_W = 640
    
    src = np.float32([[0, 0],[0, IMAGE_H], [IMAGE_W, IMAGE_H], [IMAGE_W, 0]])
    dst = np.float32([[0, 0],[int(0.34375*IMAGE_W), 230], [int(0.65625*IMAGE_W), 230], [IMAGE_W, 0]])
    M = cv2.getPerspectiveTransform(src, dst) # The transformation matrix
    
    img = img[250:IMAGE_H, 0:IMAGE_W] # Apply np slicing for ROI crop
    warped_img = cv2.warpPerspective(img,M , (IMAGE_W, IMAGE_H)) # Image warping  
    return warped_img

def sketch2(img):
    IMAGE_H = 640
    IMAGE_W = 680
    
    src = np.float32([[0, 0],[0, IMAGE_H], [IMAGE_W, IMAGE_H], [IMAGE_W, 0]])
    dst = np.float32([[0, 0],[int(0.34375*IMAGE_W), 225], [int(0.65625*IMAGE_W), 225], [IMAGE_W, 0]])
    M = cv2.getPerspectiveTransform(src, dst) # The transformation matrix
    
    img = img[250:IMAGE_H, 0:IMAGE_W] # Apply np slicing for ROI crop
    warped_img = cv2.warpPerspective(img,M , (IMAGE_W, IMAGE_H)) # Image warping  
    return warped_img

img = cv2.imread('images/car.jpg')
img = resize(img)
img = sketch(img)
#img = rotate90counterclockwise(img)

cap = cv2.VideoCapture(0)#truoc
cap1 = cv2.VideoCapture(4)#trai
cap2 = cv2.VideoCapture(2)#phai 
cap3 = cv2.VideoCapture(3)#sau
while(True):
    ret, frame = cap.read()
    ret1, frame1 = cap1.read()
    ret2, frame2 = cap2.read()
    ret3, frame3 = cap3.read()
    #front
    frame = resize(frame)
    img1 = sketch1(frame)
    
    #left side
    frame1 = resize1(frame1)
    img2 = sketch2(frame1)
    img2 = rotate90counterclockwise(img2)
    
    #right side
    frame2 = resize1(frame2)
    img3 = sketch2(frame2)
    img3 = rotate90clockwise(img3)
    
    #behind
    frame3 = resize(frame3)
    img4 = sketch1(frame3)
    img4 = rotate180(img4)
    
    
    img1 = cv2.add(img1,img)
    img1 = cv2.add(img1,img2)
    img1 = cv2.add(img1,img3)
    img1 = cv2.add(img1,img4)
    
    cv2.imshow('frame',frame2)
    cv2.imshow('sketch',img1)
    if cv2.waitKey(1) == 13:
        break
        
cap.release() 
cap1.release()
cap2.release()
cap3.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import math

def resize(img):
    IMAGE_H = 680
    IMAGE_W = 600
    dim = (IMAGE_W, IMAGE_H)
    resized = cv2.resize(img, dim)
    return resized

def resize1(img):
    IMAGE_H = 600
    IMAGE_W = 680
    dim = (IMAGE_W, IMAGE_H)
    resized = cv2.resize(img, dim)
    return resized

def rotate180(img):
    img_rotate_180 = cv2.rotate(img, cv2.ROTATE_180)
    return img_rotate_180

def rotate90counterclockwise(img):
    img_rotate_90_counterclockwise = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
    return img_rotate_90_counterclockwise

def rotate90clockwise(img):
    img_rotate_90_clockwise = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
    return img_rotate_90_clockwise

#frame   
def cut(img):
    IMAGE_H = 680
    img = img[250:IMAGE_H]
    return img
def sketch1(img):#transform large to small
    IMAGE_H = 680
    IMAGE_W = 600
    
    src = np.float32([[0, 0],[0, IMAGE_H], [IMAGE_W, IMAGE_H], [IMAGE_W, 0]])
    dst = np.float32([[0, 0],[int(0.33*IMAGE_W), 200], [int(0.67*IMAGE_W), 200], [IMAGE_W, 0]])
    M = cv2.getPerspectiveTransform(src, dst) # The transformation matrix
    
    #img = img[50:IMAGE_H] # Apply np slicing for ROI crop
    warped_img = cv2.warpPerspective(img,M , (IMAGE_W, IMAGE_H)) # Image warping  
    return warped_img
def sketch2(img):#transform small to large
    IMAGE_H = img.shape[0]
    IMAGE_W = img.shape[1]
   
    src = np.float32([[0, 0],[0, IMAGE_H], [IMAGE_W, IMAGE_H], [IMAGE_W, 0]])
    dst = np.float32([[0, 0],[-265, IMAGE_H], [IMAGE_W+270, IMAGE_H], [IMAGE_W,0]])
    M = cv2.getPerspectiveTransform(src, dst) # The transformation matrix
    
    warped_img = cv2.warpPerspective(img,M , (IMAGE_W, IMAGE_H)) # Image warping  
    return warped_img

#frame1
def cut1(img):
    IMAGE_H = img.shape[0]
    IMAGE_W = img.shape[1]
    img = img[0:IMAGE_H,0:IMAGE_W-180]
    return img
def sketch3(img):#transform large to small
    IMAGE_H = img.shape[0]
    IMAGE_W = img.shape[1]
    
    src = np.float32([[0, 0],[0, IMAGE_H], [IMAGE_W, IMAGE_H], [IMAGE_W, 0]])
    dst = np.float32([[403, int(0.293*IMAGE_H)],[403, int(0.707*IMAGE_H)], [IMAGE_W, IMAGE_H], [IMAGE_W, 0]])
    M = cv2.getPerspectiveTransform(src, dst) # The transformation matrix
    
    #img = img[50:IMAGE_H] # Apply np slicing for ROI crop
    warped_img = cv2.warpPerspective(img,M , (IMAGE_W, IMAGE_H)) # Image warping  
    return warped_img
def sketch4(img):#transform small to large
    IMAGE_H = img.shape[0]
    IMAGE_W = img.shape[1]
   
    src = np.float32([[0, 0],[0, IMAGE_H], [IMAGE_W, IMAGE_H], [IMAGE_W, 0]])
    dst = np.float32([[0, -230],[0, IMAGE_H+220], [IMAGE_W, IMAGE_H], [IMAGE_W,0]])
    M = cv2.getPerspectiveTransform(src, dst) # The transformation matrix
    
    warped_img = cv2.warpPerspective(img,M , (IMAGE_W, IMAGE_H)) # Image warping  
    return warped_img

def fix(img):
    rows = img.shape[0]
    cols = img.shape[1]
    for i in range(rows):
        for j in range(cols):
            offset_x = 0
            offset_y = int(40 * math.sin(2 * 3.14 * j / (2*cols)))
            if i+offset_y < rows:
                img[i,j] = img[(i+offset_y)%rows,j]
            else:
                img[i,j] = 0
                
    return img

#camera
cap = cv2.VideoCapture(1)#trước
cap1 = cv2.VideoCapture(2)#phải
while(True):
    #frame
    ret, frame = cap.read()
    frame = resize(frame)
    #cv2.imshow('origin',frame)
    frame = cut(frame)
    frame = sketch2(frame)
    frame = resize(frame)
    frame = sketch1(frame)
    #cv2.imshow('frame',frame)
    
    #frame1
    ret1, frame1 = cap1.read()
    frame1 = rotate90clockwise(frame1)
    frame1 = resize(frame1)
    #cv2.imshow('origin1',frame1)
    frame1 = cut1(frame1)
    frame1 = resize(frame1)

    frame1 = sketch4(frame1)
    frame1 = sketch3(frame1)
    
    #total
    #frame1 = cv2.add(frame,frame1)
    frame1 = cv2.addWeighted(frame,1.9,frame1,1.6,-80)
    cv2.imshow('frame1',frame1)
    if cv2.waitKey(1) == 13:
        break
        
    
#print(frame1.shape)       
cap.release() 
cap1.release()
cv2.destroyAllWindows()
```
